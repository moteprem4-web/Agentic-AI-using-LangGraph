```python
class BlogState(TypedDict):
    title:str
    outline:str
    content:str
```

```python
def create_outlier(state):
    #fetch title
    title=state['title']

    #call the llm gen outline
    prompt=f'Generate an outlier for a  blog topic {title}'
    outline=model.invoke(prompt).content


    state['outline']=outline

    return state
```

```python
def create_blog(state):
    title = state['title']
    outline = state['outline']

    prompt = f"write a detailed blog on title using following outline {outline}"

    content = model.invoke(prompt).content
    state["content"] = content

    return state  # type: ignore
```

```python
graph=StateGraph(BlogState)

#nodes
graph.add_node('create_outlier',create_outlier)
graph.add_node('create_blog',create_blog)

graph.add_edge(START,'create_outlier')
graph.add_edge('create_outlier','create_blog')
graph.add_edge('create_blog',END)

#
workflow=graph.compile()
```

```python
initial_state={'title':'Rise of AI in India'}

final_state=workflow.invoke(initial_state)


print(final_state['outline'])

```
